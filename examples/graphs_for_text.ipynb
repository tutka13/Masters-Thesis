```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import math
#from matplotlib.patches import Rectangle
```

```python
# Define the colormap transitioning from red to blue
colors = [(1, 0, 1), (0, 0, 0)]
cmap = mcolors.LinearSegmentedColormap.from_list('my_colormap', colors)
```

```python
# Create the data
axlimx = 2.1
axlimy = 2.1
x = np.linspace(-axlimx, axlimx, 100)
y = np.linspace(-axlimy, axlimy, 100)
X, Y = np.meshgrid(x, y)

# Latex style in text
plt.rcParams['text.usetex'] = True
```

```python
axlimx = 2.1
axlimy = 2.1
x = np.linspace(-axlimx, axlimx, 100)
y = np.linspace(-axlimy, axlimy, 100)
X, Y = np.meshgrid(x, y)

for t in np.arange(-1,1,0.1):
    Fxy =  Y - 2*X*t + t**2
    plt.contour(X, Y, Fxy, levels=[0], colors=[cmap((t+1)/2)])

Exy = Y - X**2
plt.contour(X, Y, Exy, levels=[0], colors='blue', linewidth=2)
plt.gca().set_aspect('equal')
plt.xlabel('$x$', fontsize='12')
plt.ylabel('$y$', fontsize='12')
plt.title(r'$F(x, y, t) = y - 2xt + t^2$', fontsize='12')

plt.savefig("C:\\Users\\tutko\\Desktop\\images\\odr.pdf")
```
